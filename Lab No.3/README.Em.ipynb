# Lab 3 Series: Time Series Data Preparation & Feature Engineering  
**American Electric Power (AEP) Hourly Energy Consumption Dataset**



## Overview
This repository contains the **Lab 3** series for the Machine Learning course. The labs focus on preparing and enhancing the AEP hourly energy consumption dataset through systematic data cleaning, missing value handling, and feature engineering.

### Objectives
- Handle missing timestamps in time-series data
- Perform data cleaning and validation
- Introduce external features (holidays) to improve model readiness
